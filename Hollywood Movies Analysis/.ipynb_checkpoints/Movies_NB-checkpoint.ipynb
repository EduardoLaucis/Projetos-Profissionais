```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
# Nome do arquivo local
file_name = "Most Profitable Movies of All Time.xlsx"

# Carregar todas as planilhas em um dicionário de DataFrames
sheets = pd.read_excel(file_name, sheet_name=None, engine="openpyxl")

# Separar cada planilha em uma variável
df_profitable_movies = sheets["Most Profitable Movie of All Ti"]
df_inflation = sheets["Inflation Adjusted"]
df_top_500 = sheets["Top 500 Movies (pink = added by"]

# Exibir as primeiras linhas de cada DataFrame
print("Most Profitable Movies:")
display(df_profitable_movies.head())

print("\nInflation Data:")
display(df_inflation.head())

print("\nTop 500 Movies:")
display(df_top_500.head())
```

```python
print("Tabela: df_profitable_movies\n" , df_profitable_movies.dtypes)
print("Tabela: df_inflation\n" , df_inflation.dtypes)
print("Tabela: df_top_500\n" ,df_top_500.dtypes)
```

```python
# Configuração de estilo
sns.set_theme(style="darkgrid")
```

```python
# 1. Quais os filmes mais lucrativos?

# Definindo o estilo
sns.set(style="whitegrid")

# Selecionando os 10 filmes mais lucrativos ajustados pela inflação
top_10_filmes = df_profitable_movies[['title', 'inflation adjusted gross ($m)']].sort_values(by='inflation adjusted gross ($m)', ascending=False).head(10)

# Criando o gráfico de barras horizontais
plt.figure(figsize=(12, 8))
bars = plt.barh(top_10_filmes['title'], top_10_filmes['inflation adjusted gross ($m)'], color='teal')

# Adicionando título e rótulos aos eixos
plt.title('Top 10 Filmes Mais Lucrativos Ajustados pela Inflação', fontsize=16, fontweight='bold')
plt.xlabel('Lucro Ajustado pela Inflação (em bilhões de dólares)', fontsize=12)
plt.ylabel('Título do Filme', fontsize=12)

# Função para formatar os valores em dólares (Número.Decimal Bi)
def format_currency(x, pos):
    return f'${x:,.2f}M'  # Formato: $ número com 2 casas decimais e "M" para milhões

# Configurar o formato do eixo X
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_currency))

# Adicionando os rótulos de dados nas barras
for bar in bars:
    yval = bar.get_width()
    plt.text(yval, bar.get_y() + bar.get_height() / 2, f'${yval:,.2f}M', ha='left', va='center', fontsize=10, color='black')

# Melhorando a exibição do gráfico
plt.tight_layout()
plt.show()
```

```python
# 2. Qual maior período de inflação?
plt.figure(figsize=(12, 5))
sns.lineplot(data=df_inflation, x='year', y='inflation adjusted gross ($m)', marker='o')
plt.title("Evolução da Inflação Ajustada ao Longo dos Anos")
plt.ylabel("Receita Ajustada pela Inflação ($ milhões)")
plt.xlabel("Ano")
plt.show()
```
